### Function

```
def fuzzy_merge(df1, df2, key1, key2, threshold=90, limit=2):
    """
    df_1 is the left table to join
    df_2 is the right table to join
    key1 is the key column of the left table
    key2 is the key column of the right table
    threshold is how close the matches should be to return a match, based on Levenshtein distance
    limit is the amount of matches that will get returned, these are sorted high to low
    """
    s = df2[key2].tolist()

    m = df1[key1].apply(lambda x: process.extract(x, s, limit=limit))    
    df1['matches'] = m

    m2 = df1['matches'].apply(lambda x: ', '.join([i[0] for i in x if i[1] >= threshold]))
    df1['matches'] = m2

    return df1
```

### Fuzzy

```
df3 = fuzzy_merge(df1, df2, 'key1','key2', threshold=80)
df3
df3 = pd.merge(df3, df2, how='left', left_on='matches', right_on='key2')
df3
```
